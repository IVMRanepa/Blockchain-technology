### Предлагаемая аналитическая модель

Для оценки устойчивости цифровой платформы предлагается использовать интегральный индекс:

\[
I = 0.30S + 0.25D + 0.20A + 0.15K + 0.10E
\]

где:

- \(S\) — индекс импортонезависимости программного стека;
- \(D\) — индекс доступности и качества данных;
- \(A\) — индекс зрелости аналитики;
- \(K\) — индекс кадровой обеспеченности;
- \(E\) — индекс экономической устойчивости.

Чем выше значение \(I\), тем выше способность платформы адаптироваться к санкционному давлению. Такая модель удобна для мониторинга, сравнительного анализа и сценарного прогнозирования.

### Демонстрационное решение на Python 3

Ниже приведен пример расчета интегрального индекса на **сгенерированных данных**. Этот фрагмент можно включить в статью как демонстрационный.


```python
import pandas as pd # Импортируем библиотеку pandas для работы с табличными данными (DataFrame)
import numpy as np # Импортируем библиотеку numpy для выполнения численных операций

np.random.seed(42) # Устанавливаем "зерно" для генератора случайных чисел numpy для воспроизводимости результатов

# Определяем список названий цифровых платформ
platforms = ['Маркетплейс-А', 'ГосПлатформа-Б', 'Финтех-В', 'Логистика-Г', 'Образование-Д']

# Создаем DataFrame (таблицу) с синтетическими данными для каждой платформы
df = pd.DataFrame({
    'platform': platforms, # Названия платформ
    'stack_independence': np.random.randint(55, 95, size=5), # Индекс импортонезависимости стека (случайные значения от 55 до 94)
    'data_quality': np.random.randint(50, 90, size=5), # Индекс качества данных (случайные значения от 50 до 89)
    'analytics_maturity': np.random.randint(45, 95, size=5), # Индекс зрелости аналитики (случайные значения от 45 до 94)
    'staff_capacity': np.random.randint(40, 85, size=5), # Индекс кадровой обеспеченности (случайные значения от 40 до 84)
    'economic_stability': np.random.randint(50, 90, size=5) # Индекс экономической устойчивости (случайные значения от 50 до 89)
})

# Рассчитываем интегральный индекс устойчивости для каждой платформы по заданной формуле с весовыми коэффициентами
df['resilience_index'] = (
    0.30 * df['stack_independence'] + # Вклад импортонезависимости стека (30%)
    0.25 * df['data_quality'] + # Вклад качества данных (25%)
    0.20 * df['analytics_maturity'] + # Вклад зрелости аналитики (20%)
    0.15 * df['staff_capacity'] + # Вклад кадровой обеспеченности (15%)
    0.10 * df['economic_stability'] # Вклад экономической устойчивости (10%)
).round(2) # Округляем результат до двух знаков после запятой

# Выводим DataFrame, отсортированный по убыванию интегрального индекса устойчивости
print(df.sort_values('resilience_index', ascending=False))
```

Комментарий к коду: в примере моделируются пять условных цифровых платформ, для каждой из которых задаются показатели по пяти направлениям. После этого вычисляется интегральный индекс устойчивости, позволяющий ранжировать платформы по степени готовности к работе в условиях санкционного давления.

### Интерпретация результатов

Если платформа имеет высокий показатель импортонезависимости, но низкое качество данных, то даже при наличии собственных серверов и отечественного ПО управленческие решения останутся слабо обоснованными. Напротив, сочетание качественных данных, зрелой аналитики и устойчивого стека дает более высокий адаптационный потенциал.

### Пример сценарного анализа

Следующий пример показывает, как может моделироваться ухудшение внешней среды.

```python
scenario_df = df.copy()
# Создаем копию исходного DataFrame для сценарного анализа

# Уменьшаем значения индексов для моделирования ухудшения внешней среды
# Индекс импортонезависимости программного стека уменьшаем на 10 пунктов
scenario_df['stack_independence'] = scenario_df['stack_independence'] - 10
# Индекс экономической устойчивости уменьшаем на 8 пунктов
scenario_df['economic_stability'] = scenario_df['economic_stability'] - 8
# Индекс кадровой обеспеченности уменьшаем на 5 пунктов
scenario_df['staff_capacity'] = scenario_df['staff_capacity'] - 5

# Пересчитываем интегральный индекс устойчивости ('stress_index') для сценарных условий
scenario_df['stress_index'] = (
    0.30 * scenario_df['stack_independence'] + # Вклад индекса импортонезависимости
    0.25 * scenario_df['data_quality'] +       # Вклад индекса доступности и качества данных (не изменился)
    0.20 * scenario_df['analytics_maturity'] + # Вклад индекса зрелости аналитики (не изменился)
    0.15 * scenario_df['staff_capacity'] +     # Вклад индекса кадровой обеспеченности
    0.10 * scenario_df['economic_stability']    # Вклад индекса экономической устойчивости
).round(2)

# Вычисляем разницу ('delta') между новым сценарным индексом ('stress_index') и исходным индексом устойчивости ('resilience_index')
scenario_df['delta'] = (scenario_df['stress_index'] - df['resilience_index']).round(2)

# Выводим платформы, их сценарный индекс устойчивости и изменение индекса, отсортированные по убыванию 'stress_index'
print(scenario_df[['platform', 'stress_index', 'delta']].sort_values('stress_index', ascending=False))
```

Комментарий к коду: в стресс-сценарии снижаются показатели технологической независимости, кадровой обеспеченности и экономической устойчивости. Это позволяет оценить чувствительность платформы к внешним ограничениям и заранее определить, какие подсистемы нуждаются в первоочередном укреплении.

### Пример работы с открытыми данными аналитических сайтов

Для практики в статье можно использовать открытые CSV-выгрузки или данные с аналитических порталов, публикующих статистику по цифровой экономике, ИТ-рынку, вакансиям, инвестициям или использованию ПО. При наличии открытого датасета код может выглядеть так:

```python
import pandas as pd # Импортируем библиотеку pandas для работы с данными
import matplotlib.pyplot as plt # Импортируем библиотеку matplotlib для построения графиков
import numpy as np # Импортируем библиотеку numpy для численных операций

# Генерируем синтетические данные для демонстрации, так как файл 'open_data_platform_metrics.csv' не найден
np.random.seed(42) # Устанавливаем "зерно" для воспроизводимости результатов
dates = pd.to_datetime(pd.date_range(start='2022-01-01', periods=24, freq='ME')) # Создаем 24 даты, с интервалом в месяц (конец месяца)
incidents = np.random.randint(10, 100, size=24) + np.sin(np.arange(24)/3) * 20 # Генерируем случайное количество инцидентов с сезонной компонентой
data = pd.DataFrame({'date': dates, 'incidents': incidents}) # Создаем DataFrame из сгенерированных данных

# Преобразуем столбец 'date' в формат datetime (если бы читали из файла, это было бы необходимо)
data['date'] = pd.to_datetime(data['date'])

# Группируем данные по месяцам и вычисляем среднее число инцидентов
monthly = data.groupby(pd.Grouper(key='date', freq='ME'))['incidents'].mean() # Группируем по концу месяца

# Создаем фигуру и оси для графика с заданным размером
plt.figure(figsize=(10, 5))

# Строим график среднего числа инцидентов по месяцам
monthly.plot()

# Устанавливаем заголовок графика
plt.title('Среднее число инцидентов по месяцам')

# Устанавливаем подпись оси X
plt.xlabel('Период')

# Устанавливаем подпись оси Y
plt.ylabel('Инциденты')

# Включаем сетку на графике для лучшей читаемости
plt.grid(True)

# Отображаем график
plt.show()
```

Комментарий к коду: пример демонстрирует базовую логику аналитики — загрузку открытых данных, агрегацию временного ряда и визуализацию динамики инцидентов. Аналогичным образом могут анализироваться задержки поставок, отказоустойчивость сервисов, изменение пользовательской активности и степень импортозамещения компонентов.
